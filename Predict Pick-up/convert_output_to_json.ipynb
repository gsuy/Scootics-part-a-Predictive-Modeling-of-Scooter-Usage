```
output_pickup_pred = pd.read_csv(r'/content/drive/My Drive/Colab Notebooks/Dataset/Output_pred.csv')
output_pickup_pred = np.array(output_pickup_pred)
output_pickup_actual = pd.read_csv(r'/content/drive/My Drive/Colab Notebooks/Dataset/Output_actual.csv')
output_pickup_actual = np.array(output_pickup_actual)
pickup_data = dict()
import json
for item in output_pickup_pred:
  grid_id = item[26]
  pickup_data[grid_id] = dict()

for item in output_pickup_pred:
  grid_date = item[0]
  grid_id = item[26]
  longitude = item[27]
  latitude = item[28]
  pickup_data[grid_id][grid_date] = dict()
  pickup_data[grid_id][grid_date]['predict_pickup'] = dict()
  pickup_data[grid_id][grid_date]['actual_pickup'] = dict()
  pickup_data[grid_id]['longitude'] = longitude
  pickup_data[grid_id]['latitude'] = latitude

for item in range(len(output_pickup_pred)):
  grid_date = output_pickup_pred[item][0]
  grid_start_day = output_pickup_pred[item][1]
  grid_id = output_pickup_pred[item][26]
  grid_hr_predict = output_pickup_pred[item][2:26]
  grid_hr_actual = output_pickup_actual[item][2:26]
  for itemm in range(len(grid_hr_predict)):
    pickup_data[grid_id][grid_date]['predict_pickup'][itemm] = int(grid_hr_predict[itemm])
    pickup_data[grid_id][grid_date]['actual_pickup'][itemm] = int(grid_hr_actual[itemm])
  pickup_data[grid_id][grid_date]['dayofweek'] = grid_start_day
path_file = r'/content/drive/My Drive/Colab Notebooks/Dataset/output_pickup_data.json' 
with open(path_file, 'w') as json_file:
    json.dump(pickup_data, json_file)
```
